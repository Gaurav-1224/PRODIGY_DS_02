# Task 2(Exploratory Data Analysis)

The datasets is obtained from Kaggle: [Titanic](https://www.kaggle.com/c/titanic/data)   
 
The dataset contains 891 rows (entries) and 12 columns    
The columns are:       

``PassengerId``: Unique identifier for each passenger.    
``Survived``: Binary variable indicating survival (1 = Survived, 0 = Did Not Survive).    
``Pclass``: Ticket class (1st, 2nd, 3rd class).     
``Name``: Passenger's name.    
``Sex``: Gender of the passenger.      
``Age``: Age of the passenger.     
``SibSp``: Number of siblings/spouses aboard.     
``Parch``: Number of parents/children aboard.     
``Ticket``: Ticket number.    
``Fare``: Passenger fare.    
``Cabin``: Cabin number.        
``Embarked``: Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)       


## Importing Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Loading the dataset

## Summary Statistics

```python
#For numerical features
print(df.describe(),"\n")

#For categorical features
print("\n",df.describe(include = "object"))
```

## Checking for missing values

```python
features_with_na = []
for features in df.columns:
    if (df[features].isnull().sum() > 0):
        features_with_na.append(features)
        print(f"Missing values in {features}: {df[features].isnull().sum()}/891")
```

## Age and Fare distribution of passengers

```python
# Plot distribution of age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Fare distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Fare'].dropna(), bins=30, kde=True)
plt.title('Fare Distribution of Passengers')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

## Univariate and Bivariate plots for Survival Count

```python
# Countplot for survival(Univariate)
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df)
plt.title('Survival Counts')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()

# Countplot for survival(Bivariate)
# Survival by gender
plt.figure(figsize=(8, 5))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right', labels=['Did Not Survive', 'Survived'])
plt.show()

# Survival by passenger class
plt.figure(figsize=(8, 5))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right', labels=['Did Not Survive', 'Survived'])
plt.show()
```

## Correlation of numerical features with each other

```python
# Heatmap for correlations (numerical features)
# Select only numerical columns (float and int types)
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Compute correlation matrix
plt.figure(figsize=(10, 6))
corr = numeric_df.corr()  # Correlation only on numerical data
sns.heatmap(corr, annot=True, cmap='Spectral', fmt='.2f')

plt.title('Correlation Heatmap')
plt.show()
```
